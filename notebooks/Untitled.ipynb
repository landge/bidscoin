```python
from pathlib import Path
import os 
import pydicom
from collections import Counter
```

```python
test_directory = Path("images")
```

```python
glob_result = test_directory.glob('**/*')
```

```python
possible_dicom_files_list = [x for x in glob_result if x.is_file()]
```

```python
len(possible_dicom_files_list)
```

```python
def make_series_description(tag_description):
    if tag_description == "CT":
        return "CT caput"
    elif tag_description == "CTA":
        return "CT angiography"
    elif tag_description == "MR":
        return "MR caput"
```

```python
for path in possible_dicom_files_list:
    ds = pydicom.dcmread(path, stop_before_pixels=False)
    if isinstance(ds, pydicom.dicomdir.DicomDir):
        dicom_dirs.append('new dicomdir')
    elif isinstance(ds, pydicom.dataset.FileDataset):
        id_list = ds.PatientID.split('_')
        ds.PatientID = id_list[0]
        ds.SeriesDescription = make_series_description(id_list[1])
        ds.save_as('output' / path.parent)
    else:
        pass
```
